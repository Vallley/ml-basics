```python
%matplotlib inline
from skimage.io import imread
import skimage
from matplotlib import pyplot as plt
import numpy as np
from kMean import *

img = imread("bird2.jpg")
plt.figure(figsize=(10, 15))
plt.axis('off')
img = skimage.img_as_float(img)
img_high = img.shape[0]
img_width = img.shape[1]
plt.imshow(img)
img = img.reshape(-1, 3)
clusters, labels, cent_history = kmeans_fit_predict(img, k=8, low=np.min(img), high=np.max(img))

img2 = np.copy(img)

for c in range(len(clusters)):
    img2[labels[:, -1] == c] = clusters[c]

img2 = img2.reshape(img_high, img_width, 3)
plt.figure(figsize=(10, 15))
plt.axis('off')
plt.imshow(img2)
```
