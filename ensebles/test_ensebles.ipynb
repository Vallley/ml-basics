```python

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score, classification_report

from ensebles import *
import pandas as pd
import numpy as np

link ='https://gitlab.7bits.it/coding-garden/coding-garden-ml-data/-/raw/master/data_churn-bigml-80.csv'
df = pd.read_csv(link, sep=',')

df = df.replace({
    True: 1,
    False: 0,
    'Yes': 1,
    'No': 0
})

df = df.select_dtypes(include=[np.number]).dropna()

data = df.to_numpy(dtype=float)

def decide_class(logits, threshold=0.5):
    probs = 1 / (1 + np.exp(-logits))

    return (probs >= threshold).astype(int)
train_x, test_x, train_y, test_y = train_test_split(data[:,:-1], data[:,-1], test_size=0.3, random_state=42)

tree = GradientTreeBoosting(loss='lq', iterations=20, max_depth=5, random_state=42)
tree.fit(train_x, train_y)
y = tree.predict(test_x)
y_pred = decide_class(y)

print(classification_report(test_y, y_pred))


```
